### 특정문자(열)에 접근하는 방법
- 인덱싱(Indexing) : 무엇인가를 '가리킨다'는 의미 - 하나의 인덱스 값
- 슬라이싱(Slicing) : 무엇인가를 '잘라낸다'는 의미 - 일정 범위
- 파이썬에서는 숫자를 0부터 세는 '제로 인덱스'이다

```python
#문자 인덱싱
s = 'My name is ES'
# i 접근
s[8]
# e 접근
print(s[6])
print(s[-7])
```

```python
# 문자열 슬라이싱
s[0:2] 
#name 출력
print(s[3:7])
#ES 출력
#print(s[11:13])
print(s[11:]) #처음부터 끝까지
#모든 문자열 출력
print(s[:])
```

```python
day = "2020년 3월 3일의 날씨는 맑음입니다."
print("날짜: ",day[:11])
print("날씨: ",day[-6:-4])
```

### 문자열 관련 함수

```python
# day 문자열에서 인덱스 번호 확인하기
# find 함수 활용
a = day.find('맑음') #17
day[a:a+2]
```

```python
# 문자열에서 인덱스 번호를 확인
# index 함수
day.index('맑음')
```

```python
# find, index 함수의 차이 비교 - 문장 안에 없는 문자열을 찾으려고 할 때
day.find('이경근') #-1 반환 : 없음
# day.index('이경근') # 문자를 찾을 수 없다면 에러 발생
```

```python
# len() : 문자열의 개수 반환(길이 반환)
len(day) #특수문자, 빈칸 포함
```

```python
# replace('찾을 문자열', '변경할 문자열') : 특정 문자열을 찾아서 변경하는 기능
# 맑음 -> 흐림
day = day.replace('맑음', '흐림') #데이터 추가, 삭제 등 변경할 때는 재대입 필요
```

```python
day
```

```python
# 공백 제거해주는 함수
# strip() : 양쪽 공백 제거
day2 = '  오늘은 맑습니다  '

#day2.strip()
# lstrip() : 왼쪽 공백제거
# rstrip() : 오른쪽 공백제거
day2.rstrip()
```

```python
# count('찾을 문자열') : 찾을 문자열이 몇 개 있는지 개수 반환
s.count('e')
```

```python
# 대소문자 변환
# 대문자 >> 소문자 lower()
# 소문자 >> 대문자 upper()

s.lower()
s.upper()
```

```python
# split() : 문자열 나누기
s.split()

s2 ='안녕. 점심에는. 돈까스. 먹을거야.'
s2.split('.')
```
