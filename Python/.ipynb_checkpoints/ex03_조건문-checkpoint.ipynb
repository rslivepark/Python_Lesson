### 조건문
- 상황에 따라 실행 흐름을 다르게 하는 문법

### if문
- if 조건식:
- (들여쓰기) 실행문장 //tab, space*4

```python
if 5>3:
    print('실행문장')
```

```python
if 5<3:
    print('실행문장') # 조건식이 false => 실행되지 않음
```

```python
# 들여쓰기의 중요성
if 5 < 3:
    print("실행문장")
    print("if문 안에 있는 문장")
print("if문 밖에 있는 문장")
```

```python
# 조건문을 사용하여 money변수가 10000원이상이면 '택시를 탄다' 문장 출력
money = 17000
if money >= 10000:
    print('택시를 탄다')
```

```python
# if else
money = 11000
if money >= 10000:
    print('택시를 탄다')
else:
    print('버스를 탄다')
```

```python
num1 = int(input('첫번째 정수 입력>> '))
num2 = int(input('두번째 정수 입력>> '))

if num1 > num2:
    print('첫번째 정수가 더 큽니다')
elif num1 < num2:
    print('두번째 정수가 더 큽니다')
else:
    print('두 수가 똑같습니다.')
```

```python
# 학점만들기
# 1.score라는 변수에 점수를 입력
score = int(input('점수 입력>> '))

# 2.학점판별
if score>=90:
    grade = 'A'
elif score>=80:
    grade = 'B'
elif score>=70:
    grade = 'C'
elif score>=60:
    grade = 'D'
else:
     grade = 'F'
    
print(f'{score} 점은 {grade}학점 입니다.')
```

```python
# 숙제
print('''
===========================================
================--자판기--=================
===========================================
==1.콜라 == 2.사이다 == 3.물 =====insert===
==-------==----------==------=====------===
==  600  ==    800   == 1000 ===== 000원 ==
================================== ----- ==
===========================================
==+++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++++++++++++++++++++++++++==
==+++++++++++                ++++++++++++==
==+++++++++++      SMHRD     ++++++++++++==
==+++++++++++                ++++++++++++==
==+++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++++++++++++++++++++++++++==
===========================================
===========================================
''')

# 1. 넣은 돈 입력받기
# 2. 메뉴 선택 입력받기
# 3. 잔액계산 -> 내가 선택한 메뉴에 따라 금액이 달라짐
# 4. 거스름돈을 각각 어떤 화폐로 줘야하는지 계산 후 출력

# 추가예제
# 1. 1 또는 '콜라'라고 입력해도 작동하게 만들어보기
# 2. 1(콜라), 2(사이다), 3(물) 이외의 글자를 입력하게 되면 '잘못 입력하셨습니다' 출력
# 3. 내가 먹고 싶은 금액보다 적은 돈을 insert하면 '잔액이 부족합니다' 출력
```

```python
coke = 600
sprite = 800
water = 1000

coin = int(input('insert coin >> '))
menu = input('메뉴 선택 >> ')

if menu == '1' or menu == '콜라':
    price = coke
elif menu == '2' or menu == '사이다':
    price = sprite
elif menu == '3' or menu == '물':
    price = water
else:
    print('잘못 입력하셨습니다.')

if coin >= price:
    num1000 = (coin - price) // 1000
    num500 = ((coin - price) % 1000) // 500
    num100 = (((coin - price) % 1000) % 500) // 100
    print(f"잔돈 >> 1000원 {num1000}개, 500원 {num500}개, 100원 {num100}개")
else:
    print('잔액이 부족합니다.')
```
