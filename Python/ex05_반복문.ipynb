### while문
- 반복횟수가 명확하지 않을 때 사용

```python
# 파이썬 최고 13번 출력해보기
num = 0 # 0~12까지 while문 반복
while num < 13:
    num += 1 # num = num + 1
    print("파이썬 최고")
```

```python
# 무한루프 만들기
while True:
    print('무한루프')
# 멈추는 키워드
    break
```

```python
# 두 개의 정수를 입력받아서 더하는 코드 작성하기 
# 두 개의 정수가 0이 들어올 때까지 반복

while True:
    a = int(input("첫번째 정수 입력>> "))
    b = int(input("두번째 정수 입력>> "))
    add = a + b
    print(f'두 정수의 합 : {add}')
    if (a == 0 and b== 0):
        print("프로그램이 종료되었습니다.")
        break
```

```python
# 1.현재 몸무게(now), 목표 몸무게(goal) 입력받기
now = int(input("현재 몸무게 : "))
goal = int(input("목표 몸무게 : "))
num = 0;
# 2. while 무한루프로 설정
while True:
# 2-1.감량 몸무게를 사용자로부터 입력받기
    num += 1
    lose = int(input(f"{num}주차 감량 몸무게 : "))
# 2-2. now = now - lose
    now -= lose
    if(now <= goal):
        print(f'{now}kg 달성!! 축하합니다!!')
        break;
    
```

```python
# Up, Down 게임
# 랜덤으로 뽑은 1부터 50사이의 숫자를 뽑으면 맞추는 Up, Down 게임
import random # 함수, 모듈 가져오는 keyword
ran_num = random.randint(1, 50)

while True:
    input_num = int(input("숫자를 입력하세요 >> "))
    if(ran_num == input_num):
        print("정답 맞춤 커피 쿠폰 ㅊㅊ")
        break;
    elif(ran_num > input_num):
        print("up")
    elif(ran_num < input_num):
        print("down")
```

### for loops
- 반복횟수가 정해져 있을 경우

```python
list_food = ['국밥','샌드위치','피자','짜장면']

for food in list_food:
    print(food)
```

```python
# 문자열 반복문
hi = '안녕하세요'

for s in hi:
    print(s, end = "")
```

```python
# 튜플 반복문
tuple_icecream = ('바밤바', '누가바', '메로나', '죠스바', '쌍쌍바')
for icecream in tuple_icecream:
    print(icecream, end = " ")
```

```python
# 정보처리기사 자격증 시험 점수가 합격 or 불합격인지 판별
score_list = [90, 45, 70, 60, 55]

for i in score_list:
    if(i>=60):
        print(f"{score_list.index(i) + 1}번 학생은 합격입니다.")
    else:
        print(f"{score_list.index(i) + 1}번 학생은 불합격입니다.")
```

### for - range() 구문

```python
# range() : 필요한만큼의 숫자를 만들어내는 기능
# range(시작할 숫자, 종료할 숫자, 증감량) # 증감량은 1이 default
for i in range(10):
    print(i + 1, end = " ")
```

```python
# 1부터 99까지 3씩 증가하는 숫자 출력
for i in range(1, 100, 3):
    print(i, end = " ")
```

```python
# 10~2까지 1씩 감소하며 숫자생성
for i in range(10, 1, -1):
    print(i, end = " ")
# for i in range(11, 2, -1):
    # print(i-1, end = " ")
```

```python
# 시작값 생략 : 시작값을 생략하면 0부터 시작함
for i in range(10):  # 0~9까지 1씩 증가 후 출력
    print(i)
```

```python
# range() 함수 활용해서 요소 출력하기
list_food

for i in range(len(list_food)):
    print(i, list_food[i], end = "/")
```
