### Matplotlib
- 2D 그래프로 시각화가 가능한 라이브러리
- 파이썬에서 데이터를 차트로 그려줌
- 차트 종류 : 산점도, 바, 파이, 히스토그램, 꺾은선 그래프 등이 있음

### 데이터 시각화란?
- 광범위하게 분산된 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트 등으로 정리하는 것
- 시각화를 통해서 데이터의 특징을 쉽게 파악할 수 있음

```python
# 'matplotlib' 라이브러리에서 'pyplot'을 불러오고 'plt'라는 별칭부여
import matplotlib.pyplot as plt
```

### 기본적인 선 그래프 그려보기

```python
# Line Plot(선 그래프)
y = [2,12,6]

# 그래프를 그려주는 역할
# 데이터를 하나의 리스트만 전달하면 x축은 인덱스를 자동부여
plt.plot(y)

# 그동안 그렸던 plot 출력하는 역할
# print()같은 기능
plt.show() # 다 그렸다면 show 출력해주기!!!!!!
```

```python
# 데이터 2개 넣어보기
# x축, y축 설정해서 그려보기


x = [1,2,3,]
y = [2,4,6]


# 선 스타일을 설정해보자
plt.plot(x, y, ls = '-.', color = 'gray', marker = 'v' )

plt.savefig('line_plot.png') #이미지로 저장하고 싶을 때 사용
plt.show()
```
