### reshape()
- 원래있던 배열 새로운 차원의 배열로 재배치할 때 사용
- np.reshape(변경할 배열의 shape)
<BR><BR>
- print() #함수
- np.arange(1,51) # 메소드
- arr1.shape # 속성
<br> method chaining

```python
#1부터50까지 1씩 증가하는 2차원 넘파이 배열(5, 10) 생성
arr20 = np.arange(1, 51).reshape(5,10)
arr20
```

```python
# reshape 사용시 행과 열 둘 중에 하나의 값만 정해주고 나머지 값은 -1로 지정하면 자동으로 재배치
np.arange(1, 51).reshape(5,-1)
```

```python
# 13 가져오기
print(arr20[1,2])

#15,16,17 가져오기
print(arr20[1, 4:7])

# 3행의 전체열 가져오기
print(arr20[3, : ])

# 전체행, 0열만 가져오기
print(arr20[0:, 0:1])
print(arr20[ : , 0])

# 22. 33 데이터 가져오기
print(arr20[2,1], arr20[3,2])
print(arr20[[2, 3],[1,2]])
```

```python
# 24(2,3), 19(1,-2), 39(3,-2) 데이터 가져오기
print(arr20[[2, 1,3],[3,-2,-2]])
print(arr20[[2, 1,3],[3,8,8]])
```

### BMI 구하기 실습

#### txt 파일 불러오기 <BR>
np.loadtxt('height_weight(p).txt', delimiter = ',')
   
#### delimiter : 데이터가 어떤 기호를 기준으로 구분되었는지 알려줌(구분문자)

```python
data = np.loadtxt('height_weight(p).txt', delimiter = ',')
data 
```

```python
# 데이터의 크기 확인
# 2행 10열 2차원 넘파이 배열
data.shape
```

```python
# BMI 구하는 공식 
# 몸무게(kg) / 키(m) *키(m)
# 단위변환후 -> BMI 구하기
```

```python
# 1. 키 데이터 가져오기
print(data[0, :])

# 2. 키데이터 단위 변환 cm -> m (m 변수에 담기) // 현재 키데이터 * 0.01
m = data[0, :] * 0.01
print(m)

# 3. 몸무게 데이터 가져오기
print(data[1, :])

# 4. 몸무게 데이터 단위 변환 pound>> kg (kg 변수에 담기) // pound/2.204
kg = data[1, :] / 2.204
print(kg)

# 5. BMI 구하기
bmi = kg / (m*m)
print(np.round(bmi, 2))
```

### boolean 인덱싱(색인)/마스킹 (masking)
- 필터링의 기능을 담당
- 조건에 해당하는 (True) 값들만 출력할 때 사용

```python
array1d = np.arange(1, 10, 1)
array1d
```

```python
# 현재는 논리연산만 진행한 결과임ㅇㅇ 인덱싱 아님!!!!!
array1d > 5
```

```python
# 배열명[조건]
array1d[array1d > 5]
```

```python
# 점수로 불리언 인덱싱 진행
score = np.array([80, 75, 55, 96, 30])
score
```

```python
# 점수가 80점 이상인 경우만 출력
score>=80 # 논리 연산만 진행한 결과 
score[score>=80]
```

```python
name = np.array(['종찬', '경근', '현민', '고봉'])
name
```

```python
# 불리언 인덱싱으로 이름이 '종찬'인 사람만 가져오기
print(name[name == '종찬'])
```

```python
score2 = np.array([[60,60], [70,80], [90,20], [50,50]])
score2
```

```python
score2[name == '종찬']
```

### numpy 범용함수
- 범용함수란? 널리 퍼져있는 수학 함수
- sum(), mean(), max(), min()

```python
# sum() : 합계를 구해주는 메소드
# arr라는 변수에 랜덤수 (1~10) 들어있는 2행 5열인 2차원 배열 생성
arr = np.random.randint(1, 11, (2,5))
arr
```

```python
# np.sum(배열명)
print(np.sum(arr))

# 평균
print(np.mean(arr))

# 최댓값
print(np.max(arr))
```

```python
# 절대값을 구해주는 메소드 : abs()
arr_abs = np.array([-1,-2,-3,-4,-5])
print(np.abs(arr_abs))
```
